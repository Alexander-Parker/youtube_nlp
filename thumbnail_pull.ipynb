```python
#vid search function

import json
import os
import boto3
import time

from apiclient.discovery import build  
#from apiclient.discovery import build



service = build('youtube', 'v3', developerKey='YOUR KEY HERE')

search_response = service.search().list(
    q='donald trump rally',
    type="video",
    part="snippet",
    maxResults=50
).execute()
```

```python
url_list = []

for item in search_response['items']:
    url_list.append(item['snippet']['thumbnails']['high']['url'])
```

```python
from PIL import Image
import requests
from io import BytesIO
import urllib.request


image_list = []

for url in url_list:
    urllib.request.urlretrieve(url, 'thumbnails/' + url.split('/')[4] + '.jpg')
```
